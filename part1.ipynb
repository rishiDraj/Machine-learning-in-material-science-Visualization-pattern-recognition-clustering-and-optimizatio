 co-variance of dataset

```python
print(data)
x1= data["Atomic Number"]
x2= data["Electronegativity"]
x3= data["Atomic Radius"]
x4= data["Thermal Conductivity"]
x5= data["Density"]
```

```python
#A function to standardize the data w.r.t zero mean

def standardize(X): # X is the input data
    mu=sum(X)/len(X)
    var=sum((X-mu)**2)/len(X)
    z=(X-mu)/(var**0.5)
    return z
```

```python
A= [x1,x2,x3,x4,x5]
B= np.array(A)
B=B.T
C= standardize(B)
```

```python
D= np.dot(C.T,C)/94
```

```python
pd.DataFrame(D)
```

#### We can see that the most covariance is between Atomic number and density
